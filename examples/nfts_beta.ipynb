```python
import keyring
from pycgapi import CoinGeckoAPI
```

```python
api_key = keyring.get_password('coingecko', 'analyst_api_key')
```

```python
api = CoinGeckoAPI(api_key, pro_api=True)
status = api.status_check()
print("API Server Status:", status)
```

## /nfts/list

```python
supported_nfts = api.nfts_supported()
supported_nfts.head(10)
```

## /nfts/{id}

```python
nft_collection_data = api.nft_collection_info(nft_id='bored-ape-yacht-club')
nft_collection_data
```

```python
nft_ids = [
    'galxe-oat-v2', 
    'cryptopunks', 
    'cometh-spaceships', 
    'bored-ape-yacht-club', 
    'yuliorigingenone'
]
nft_collections = api.nft_collections_info(nft_ids)
nft_collections
```
