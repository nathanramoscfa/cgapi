```python
import keyring
from pycgapi import CoinGeckoAPI
```

```python
api_key = keyring.get_password('coingecko', 'analyst_api_key')
```

```python
api = CoinGeckoAPI(api_key, pro_api=True)
status = api.status_check()
print("API Server Status:", status)
```

## /coins/list/new

```python
latest_coins_listed = api.new_coins_listed()
latest_coins_listed.head()
```

## /coins/top_gainers_losers

```python
gainers, losers = api.gainers_losers()
```

```python
gainers.head()
```

```python
losers.head()
```

## /global/market_cap_chart

```python
market_cap_df, volume_df = api.historical_global_market_cap()
```

```python
market_cap_df
```

```python
volume_df
```

## /nfts/markets

```python
get_nft_markets = api.nft_market_data()
get_nft_markets.head()
```

## /nfts/{id}/market_chart

```python
nft_historical_data = api.nft_historical_data(nft_id='bored-ape-yacht-club')
nft_historical_data
```

## /nfts/{asset_platform_id}/contract/{contract_address}/market_chart

```python
nft_historical_data_by_contract = api.nft_historical_data(
    asset_platform_id='ethereum', 
    contract_address='0xbc4ca0eda7647a8ab7c2061c2e118a18a936f13d',
    days=7
)
nft_historical_data_by_contract
```
